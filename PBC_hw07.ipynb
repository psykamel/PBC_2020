```python
# 第一題較簡潔的作法
# 載入套件
import operator

# 輸入資料：文件路徑、關鍵字
filepath = input()
keyword = input()

# 讀取文件與前處理
file = open(filepath, 'rt', encoding="utf-8")
text = file.read()
text = text.replace('\t', '\n')
text = text.split('\n')

# 建立字典
former_words = dict()
latter_words = dict()
for sentence in text:
    sentence = sentence.strip()
    key_index = 0
    # searching for every keyword in the sentence
    while sentence.find(keyword, key_index) != -1:
        key_index = sentence.find(keyword, key_index)
        # append former_words
        former_word_index = key_index - 1
        if former_word_index >= 0:
            former_word = sentence[former_word_index]
            if former_word in former_words:
                former_words[former_word] += 1
            else:
                former_words[former_word] = 1
        # append latter words
        latter_word_index = key_index + len(keyword)
        if latter_word_index <= len(sentence) - 1:
            latter_word = sentence[latter_word_index]
            if latter_word in latter_words:
                latter_words[latter_word] += 1
            else:
                latter_words[latter_word] = 1
        key_index += 1

# 排序字出現次數
def get_top_ten(words_dict):
    '''
    input: a dictionary of (str, count)
    fisrt sort words by the internal code of words
    then sort them by its counts
    '''
    # sort words
    sorted_words = sorted(words_dict.items(), key=operator.itemgetter(0), reverse=True)
    sorted_words = sorted(sorted_words, key=operator.itemgetter(1), reverse=True)
    # get top ten
    result_words = sorted_words[0:min(10, len(sorted_words))]

    return(result_words)

# 輸出結果
result_former_words = get_top_ten(former_words)
result_latter_words = get_top_ten(latter_words)
print('熱門前一個字:')
for popularword, _ in result_former_words:
    print('{popular_word}---{key_word}'.format(popular_word=popularword, key_word=keyword))
print('熱門下一個字:')
for popularword, _ in result_latter_words:
    print('{key_word}---{popular_word}'.format(popular_word=popularword, key_word=keyword))
```

```python
# 載入套件
import operator

# 輸入資料：文件路徑、關鍵字
filepath = input()
keyword = input()

# 讀取文件與前處理
file = open(filepath, 'rt', encoding="utf-8")
text = file.read()
text = text.replace('\t', '\n')
text = text.split('\n')

# 建立字典
former_words = dict()
latter_words = dict()
for sentence in text:
    sentence = sentence.strip()
    key_index = 0
    # searching for every keyword in the sentence
    while sentence.find(keyword, key_index) != -1:
        key_index = sentence.find(keyword, key_index)
        # append former_words
        former_word_index = key_index - 1
        if former_word_index >= 0:
            former_word = sentence[former_word_index]
            if former_word in former_words:
                former_words[former_word] += 1
            else:
                former_words[former_word] = 1
        # append latter words
        latter_word_index = key_index + len(keyword)
        if latter_word_index <= len(sentence) - 1:
            latter_word = sentence[latter_word_index]
            if latter_word in latter_words:
                latter_words[latter_word] += 1
            else:
                latter_words[latter_word] = 1
        key_index += 1

# 排序字出現次數
def get_top_ten(words_dict):
    '''
    input: a dictionary of (str, count)
    fisrt sort words by its counts
    then sort them by the internal code of words
    '''
    # sort words by its counts
    sorted_words = sorted(words_dict.items(), key=operator.itemgetter(1), reverse=True)
    # group tuples based on counts of the word
    grouped_sorted_words = []
    for count in sorted(list(set(words_dict.values())), reverse=True):
        grouped_sorted_words.append([word for word in sorted_words if word[1] == count])
    # get top ten
    i_word = 0
    result_words = []
    for tuples in grouped_sorted_words:
        tuples = sorted(tuples, reverse=True)
        for tple in tuples:
            if i_word == 10:
                break
            result_words.append(tple)
            i_word += 1

    return(result_words)

# 輸出結果
result_former_words = get_top_ten(former_words)
result_latter_words = get_top_ten(latter_words)
print('熱門前一個字:')
for i in range(len(result_former_words)):
    print('{popular_word}---{key_word}'.format(popular_word=result_former_words[i][0], key_word=keyword))
print('熱門下一個字:')
for i in range(len(result_latter_words)):
    print('{key_word}---{popular_word}'.format(popular_word=result_latter_words[i][0], key_word=keyword))
```

```python
# 輸入資料：u.item 路徑、u.genre 路徑、電影 ID
u_item_filepath = input()
u_genre_filepath = input()
movie_id = input()

# 處理u.item
with open(u_item_filepath, 'rt', encoding="ISO-8859-1") as file_items:
    items = file_items.read()
    items = items.strip()
    items = items.split('\n')
    items_dict = dict()
    for item in items:
        item = item.split('|')
        items_dict[item[0]] = [item[1], item[5:24]]  # 建立電影字典，key為電影編號，value為[電影名, 類型]

# 處理u.genre
with open(u_genre_filepath, 'rt', encoding="ISO-8859-1") as file_genres:
    genres = file_genres.read()
    genres = genres.strip()
    genres = genres.split('\n')
    genres = [tuple(genre.split('|')) for genre in genres]
    genres_dict = {genre_id: genre_name for genre_name, genre_id in genres}  # 建立類型字典，key為類型編號，value為類型名稱

# 輸出結果
if movie_id in items_dict:
    movie_genre_id = [str(i) for i, value in enumerate(items_dict[movie_id][1]) if value == '1']  # 將類型的one-hot encoding轉為類型編號
    movie_genre_name = [genres_dict[genre_id] for genre_id in movie_genre_id]  # 將類型編號轉為類型名稱
    print('{movie_name}: {movie_genre}'.format(movie_name=items_dict[movie_id][0], movie_genre=', '.join(movie_genre_name)))
else:
    print('No movie found.')
```

# 第三題

【測資6的一直錯誤】
嗨同學，

這題每回合依序各公司預期購買的張數就如同學所說，如果符合的公司數量小於每輪預期購買張數的公司數量，則依照真正符合目標類別的公司數量，在每輪預期購買張數中多餘的公司不做考慮。

不過在剩餘張數不夠購買當前目標公司的預期購買張數時，不是直接把剩下的都給分數最高的那家。在決定真實購買張數時，是一個回合一個回合累積購買張數才決定的，而每個回合中又依序由分數高到低的公司先後的決定購買張數，也就是在最後一個回合中，依照公司分數高到低決定此回合此公司的購買數量可能出現剩餘票數不夠購買此公司的預期張數，這個時候不需要湊買預期張數，剩下的張數就全部購買當前的此公司，所以不一定會是分數最高的公司。

例如，如果本題要購買 42 張，那結束三回合後，目前還有 8 張。

第一家 7＋7 ＝14

第二家 5＋5 ＝10

第三家 3＋3＝6

第四家 2＋2＝4

因為還有剩餘張數，所以繼續下一回合，也還是依序第一家公司預期購買 7 張，8 大於 7 所以此回合真正購買也是 7 張，目前剩下 1 張並依序要決定第二家公司，第二家公司的預期購買張數是 5，但是剩餘張數不足，則剩下的張數全部購買當前公司，也就是剩下的 1 張全部買第二家公司，最後購買張數就是第一家 21 張、第二家 11 張、第三家 6 張、第四家 4 張。

```python
# 資料輸入：新聞標題檔案路徑
news_title_path = input()

# 資料輸入：關鍵字辭典檔案路徑
news_dict_path = input()

# 資料輸入：公司類別檔案路徑
company_category_path = input()

# 資料輸入：有興趣的產業類別k、購買的總張數q、每輪的購買張數r
preferences = input().split(',')

industry_interested = preferences[0]  # 有興趣的產業類別k
buy_total = int(preferences[1])  # 購買的總張數q
buy_each_choice = [int(i) for i in preferences[2].split(':')]  # 每輪的購買張數r

# 建立新聞標題集合
with open(news_title_path, encoding="utf-8") as news_title_file:
    news_titles = news_title_file.read()
    news_titles = news_titles.split('\n')

# 建立關鍵字與對應權重字典
with open(news_dict_path, encoding="utf-8") as news_dict_file:
    news_dict = news_dict_file.read()  # 讀入關鍵字-權重原始檔案
    news_dict = news_dict.split('\n')  # 分行，一行為一個關鍵字-權重字串
    news_dict = [tuple(keyword_weight.split(' ')) for keyword_weight in news_dict]  # 將關鍵字-權重字串存成(關鍵字, 權重)tuple
    news_dict = {keyword: int(weight) for keyword, weight in news_dict}  # 將(關鍵字, 權重)tuple存成字典
    keyword_list = list(news_dict.keys())  # 將關鍵字存成一個list，在斷詞函數tokenization()會使用到

# 建立公司類別字典
with open(company_category_path, encoding="utf-8") as company_category_file:
    company_categories = company_category_file.read()  # 讀入公司類別檔案
    company_categories = company_categories.split('\n')  # 分行，一行為一個公司-類別字串
    company_categories = [tuple(company_category.split(' ')) for company_category in company_categories]  # 將公司-類別字串存成(公司, 類別)tuple
    category_companies_dict = dict()  # 將(公司, 類別)tuple存成字典，key為類別，value為屬於該類別的公司所組成的list
    for company, category in company_categories:
        try:
            category_companies_dict[category].append(company)
        except:
            category_companies_dict[category] = [company]


# 斷詞函數
def tokenization(text, company, keyword=keyword_list):
    '''輸入文本text，依據給定的keyword_list進行斷詞，並輸出由token組成的list'''
    # 開始標籤
    text = text.strip()
    # 步驟一：找出所有關鍵字-所在位置組合
    key_index_collection = []  # 關鍵字-所在位置組合列表
    for key in keyword:
        key_index = 0
        while text.find(key, key_index) != -1:
            key_index = text.find(key, key_index)
            key_index_collection.append([-len(key), key_index, key])
            key_index += len(key)
    key_index_collection.sort()  # 標記字的順序：依「關鍵字字數」、「所在位置」排列

    # 步驟二：找出欲標記的字
    occ_mark = [''] * len(text)  # 表徵文本在某位置的字是否已被標記，以避免重複標記
    to_slice_key_and_index = []  # 排除掉所有重複標記的情形後的關鍵字-所在位置組合
    for key_and_index in key_index_collection:
        start_index = key_and_index[1]  # 該字在原始文本中的開始位置
        end_index = key_and_index[1] + len(key_and_index[2])  # 該字在原始文本中的結束位置+1
        # 這一組關鍵字-所在位置尚未被標籤，則執行標籤
        if 'o' not in occ_mark[start_index:end_index]:
            to_slice_key_and_index.append([key_and_index[1], '/' + key_and_index[2] + '/'])
            # 標籤完，在occ_mark上標記o，代表文本中的這個位置已被標籤
            for i in range(start_index, end_index):
                occ_mark[i] = 'o'

    # 步驟三：找出沒有標記的字，並與上者融合成完整字串
    non_key_index = [i for i, x in enumerate(occ_mark) if x == '']  # 沒有標記的字的所在位置
    non_key_and_index = [[i, text[i]] for i in non_key_index]  # 依所在位置取出對應原始文本的字元
    complete_text = [key_and_index[1] for key_and_index in sorted(non_key_and_index + to_slice_key_and_index)]
    result_text = ''.join(complete_text)  # 合成完整的句子
    result_text = result_text.replace('//', '/')  # 去除重複的/
    result_text = result_text.strip('/')  # 去除句子開頭與結尾的/

    # 輸出結果
    tokens = result_text.split('/')  # 最後去除/的分隔，輸出tokens組成的list
    return(tokens)

# 執行公司分數計算與排名
if industry_interested not in category_companies_dict.keys():
    print('NO_MATCH')
else:
    # 第一步：計算各公司分數
    industry_company_weight = dict()
    # 檢視每一個標題，若裡頭有有興趣公司，則計算分數
    for news_title in news_titles:
        for company in category_companies_dict[industry_interested]:
            if company not in news_title:
                continue
            tokens = tokenization(news_title, company, keyword_list)
            # 加總標題中的關鍵字權重
            for token in tokens:
                if token not in keyword_list:
                    continue
                try:
                    industry_company_weight[company] += news_dict[token]
                except:
                    industry_company_weight[company] = news_dict[token]

    # 第二步：依分數和公司字串內碼排行公司
    sorted_industry_company_weight = sorted(list(industry_company_weight.items()), key=lambda c: c[0], reverse=True)
    sorted_industry_company_weight = sorted(sorted_industry_company_weight, key=lambda c: c[1], reverse=True)

    # 第三步：開買股票
    num_choice = len(buy_each_choice)
    num_company = len(category_companies_dict[industry_interested])
    num_combinition = min(num_choice, num_company)

    company_stock = dict()  # 建構公司-購買張數的字典
    i_choice = 0  # 要求的購買股票張數次序的編號
    i_company_picked = 0  # 公司購買順序的編號
    while buy_total > 0:
        # 若剩餘張數滿足當前要求的購買張數，則進行購買，並進行下一次序的購買
        # 若剩餘張數無法滿足當前要求的購買張數，則僅購買完剩餘張數即可，並結束整個購買流程
        if buy_total - buy_each_choice[i_choice] >= 0:
            try:
                company_stock[sorted_industry_company_weight[i_company_picked][0]] += buy_each_choice[i_choice]
            except:
                company_stock[sorted_industry_company_weight[i_company_picked][0]] = buy_each_choice[i_choice]
            finally:
                buy_total -= buy_each_choice[i_choice]
                if i_company_picked < num_combinition - 1:
                    i_company_picked += 1
                else:
                    i_company_picked = 0
        else:
            try:
                company_stock[sorted_industry_company_weight[i_company_picked][0]] += buy_total
            except:
                company_stock[sorted_industry_company_weight[i_company_picked][0]] = buy_total
            finally:
                break
        # 若完成滿足要求張數的購買，下一次的購賣會繼續看下一個次序的購買要求
        # 若一整輪的購買都完成了，會再進行新一輪的購買，故編號歸零
        if i_choice < num_combinition - 1:
            i_choice += 1
        else:
            i_choice = 0

    # 輸出結果
    sorted_company_stock = [(company, company_stock[company]) for company, _ in sorted_industry_company_weight if company in company_stock]
    for company, stock in tuple(sorted_company_stock):
        print('{company_name}購買{stock_number}張'.format(company_name=company, stock_number=stock))
```
