# 第一題

````python
# 輸入資料：關鍵字字串、內文、INPUT_END
key_word = input()
text_collection = []
while True:
    text = input()
    if text != "INPUT_END":
        text = text.strip()
        text_collection.append(text)
    else:
        break

# 整合成一個字串，並在換行處加一個space
paragraph = " ".join(text_collection)

# 執行標籤
marked_text_collection = []
if key_word in paragraph:
    arg_beg = 0
    while paragraph.find(key_word, arg_beg) != -1:
        key_word_index = paragraph.find(key_word, arg_beg)
        sliced_start_index = key_word_index-7 if key_word_index-7 >= 0 else 0
        sliced_end_index = key_word_index+len(key_word)+7 if key_word_index+len(key_word)+7 <= len(paragraph) else len(paragraph)
        sliced_start = paragraph[sliced_start_index:key_word_index]
        sliced_end = paragraph[key_word_index+len(key_word):sliced_end_index]
        sliced_text = sliced_start + '**' + key_word + '**' + sliced_end
        marked_text_collection.append(sliced_text)
        arg_beg = paragraph.find(key_word, arg_beg) + 1
else:
    marked_text_collection.append('NO_MATCH')

# 輸出結果
for marked_text in marked_text_collection:
    print(marked_text)
````

# 第二題
1. 如果在key_1的distance內有若干個key_2，要分別列出來，然後才換下一個key_1
    * key_1 搭配 key_2-1
    * key_1 搭配 key_2-2
    * ...
2. key_2的搜尋要從index_key_1 + len(key_1)之後開始，而非+1，否則遇到疊字會把key_1的一部分也視作key_2  (感謝「天天」的測資)
```python=
index_key_2 = paragraph.find(key_2, index_key_1 + len(key_1))
```

````python
# 2020/11/8
# 輸入資料：距離參數、關鍵字一、關鍵字二、內文、INPUT_END
distance = int(input())
key_1 = input()
key_2 = input()

text_collection = []
while True:
    text = input()
    if text != 'INPUT_END':
        text_collection.append(text.strip())
    else:
        break

# 整合成一個字串，並在換行處加一個space
paragraph = " ".join(text_collection)

# 執行標籤
marked_text_collection = []
index_key_1 = 0
index_key_2 = 0
while paragraph.find(key_1, index_key_1) != -1:
    index_key_1 = paragraph.find(key_1, index_key_1)
    index_key_2 = paragraph.find(key_2, index_key_1 + len(key_1))
    while paragraph.find(key_2, index_key_2) != -1:
        index_key_2 = paragraph.find(key_2, index_key_2)
        if index_key_2 - (index_key_1 + len(key_1)) <= distance:
            # get sliced index
            sliced_start_index = index_key_1-7 if index_key_1-7 >= 0 else 0
            sliced_end_index = index_key_2+len(key_2)+7 if index_key_2+len(key_2)+7 <= len(paragraph) else len(paragraph)
            # get sliced text
            sliced_start = paragraph[sliced_start_index:index_key_1]
            sliced_middle = paragraph[index_key_1+len(key_1):index_key_2]
            sliced_end = paragraph[index_key_2+len(key_2):sliced_end_index]
            # combine sliced text
            sliced_text = sliced_start + '**' + key_1 + '**' + sliced_middle + '**' + key_2 + '**' + sliced_end
            marked_text_collection.append(sliced_text)
        # search next key_2
        index_key_2 += 1
    else:
        index_key_1 += 1
else:
    if not marked_text_collection:
        marked_text_collection.append('NO_MATCH')

# 輸出結果
for marked_text in marked_text_collection:
    print(marked_text)
````

# 第三題

````python
# 輸入資料：公司名稱集合、關鍵字集合、文本、INPUT_END
company = input().split(',')
keyword = input().split(',')
text_collection = []
while True:
    text = input()
    if text != "INPUT_END":
        text_collection.append(text.replace(' ', ''))
    else:
        break

# 開始標籤
result = []  # 最終結果儲存於此
for text in text_collection:
    # 第一部分：辨認標題是否存在所關心的公司
    company_all_and_count = [[company_name, text.count(company_name)] for company_name in company]
    company_all_and_count = sorted(company_all_and_count, key=lambda l: l[1], reverse=True)
    company_present = [company_name[0] for company_name in company_all_and_count if company_name[1] != 0]
    # 若此標題至少存在一個所關心的公司，則記錄下出現的公司名，並繼續判斷，否則跳過這一次的迴圈
    if company_present:
        result_company = ','.join(company_present)
    else:
        result.append('NO_MATCH')
        continue

    # 第二部分：處理文本切割
    # 步驟一：找出所有關鍵字-所在位置組合
    key_index_collection = []  # 關鍵字-所在位置組合列表
    for key in keyword:
        key_index = 0
        while text.find(key, key_index) != -1:
            key_index = text.find(key, key_index)
            key_index_collection.append([-len(key), key_index, key])
            key_index += len(key)
    key_index_collection.sort()  # 標記字的順序：依「關鍵字字數」、「所在位置」排列

    # 步驟二：找出欲標記的字
    occ_mark = [''] * len(text)  # 表徵文本在某位置的字是否已被標記，以避免重複標記
    to_slice_key_and_index = []  # 排除掉所有重複標記的情形後的關鍵字-所在位置組合
    for key_and_index in key_index_collection:
        start_index = key_and_index[1]  # 該字在原始文本中的開始位置
        end_index = key_and_index[1] + len(key_and_index[2])  # 該字在原始文本中的結束位置+1
        # 這一組關鍵字-所在位置尚未被標籤，則執行標籤
        if 'o' not in occ_mark[start_index:end_index]:
            to_slice_key_and_index.append([key_and_index[1], '/' + key_and_index[2] + '/'])
            # 標籤完，在occ_mark上標記o，代表文本中的這個位置已被標籤
            for i in range(start_index, end_index):
                occ_mark[i] = 'o'

    # 步驟三：找出沒有標記的字，並與上者融合成完整字串
    non_key_index = [i for i, x in enumerate(occ_mark) if x == '']  # 沒有標記的字的所在位置
    non_key_and_index = [[i, text[i]] for i in non_key_index]  # 依所在位置取出對應原始文本的字元
    complete_text = [key_and_index[1] for key_and_index in sorted(non_key_and_index + to_slice_key_and_index)]
    result_text = ''.join(complete_text)  # 合成完整的句子
    result_text = result_text.replace('//', '/')  # 去除重複的/
    result_text = result_text.strip('/')  # 去除句子開頭與結尾的/

    # 第三部分：整合第一部份和第二部分的結果
    result.append(result_company + ';' + result_text)

# 輸出結果
print(*result, sep="\n")
````
