```python
# 輸入一：物品個數、負重上限
item_num_and_max_weight = input().split(',')
item_num = int(item_num_and_max_weight[0])
max_weight = int(item_num_and_max_weight[1])

# 輸入二：物品重量
item_weight = input().split(',')
item_weight = [int(i_weight) for i_weight in item_weight]

# 輸入三：物品效用
item_utility = input().split(',')
item_utility = [int(i_utility) for i_utility in item_utility]

# 建立物品編號列表
item_index = [i for i in range(1, item_num + 1)]

# 複製一份給第二個演算法用的資料
item_index_algo2 = []
item_weight_algo2 = []
item_utility_algo2 = []
for i_item in range(item_num):
    item_index_algo2.append(item_index[i_item])
    item_weight_algo2.append(item_weight[i_item])
    item_utility_algo2.append(item_utility[i_item])


# 演算法一：CP值法
item_cp = []
for i_item in range(item_num):
    item_cp.append(item_utility[i_item] / item_weight[i_item])

# 根據CP值大小，將各物品由CP值大依序排列至CP值小者
i_comparison = 0  # 初始值從CP值列表的第一個值開始
while i_comparison < item_num - 1:  # 當比較到最後一個值時則停
    # 從自己的下一個開始比較到最後一個，在所有CP值比自己還大的值中最靠近右邊的，與之交換位置(aka越小的要排到越右邊)
    index_to_exchange = i_comparison  # 交換對象的初始值為自己
    for i_following in range(i_comparison + 1, item_num):
        # 若比較對象的CP值大於自己，應與之交換，將其索引值暫存
        if item_cp[i_comparison] < item_cp[i_following]:
            index_to_exchange = i_following
        # 若比較對象的CP值等於自己的，則進一步判斷重量
        elif item_cp[i_comparison] == item_cp[i_following]:
            # 若比較對象的重量小於自己，應與之交換，將其索引值暫存
            if item_weight[i_comparison] > item_weight[i_following]:
                index_to_exchange = i_following
            # 若比較對象的重量等於自己的，則進一步判斷編號值
            elif item_weight[i_comparison] == item_weight[i_following]:
                # 若比較對象的編號比自己小，應與之交換，將其索引值暫存
                if item_index[i_comparison] > item_index[i_following]:
                    index_to_exchange = i_following
    # 當判斷出有交換時，進行位置交換
    if index_to_exchange != i_comparison:
        item_cp[i_comparison], item_cp[index_to_exchange] = item_cp[index_to_exchange], item_cp[i_comparison]
        item_utility[i_comparison], item_utility[index_to_exchange] = item_utility[index_to_exchange], item_utility[i_comparison]
        item_weight[i_comparison], item_weight[index_to_exchange] = item_weight[index_to_exchange], item_weight[i_comparison]
        item_index[i_comparison], item_index[index_to_exchange] = item_index[index_to_exchange], item_index[i_comparison]
    # 當那個位置之後的效能值沒有再比自己更大時，表這個位置的乘客已經確定，下一個loop會開始確認下一個位置
    else:
        i_comparison += 1

# 根據剛剛排出的順序，依序判斷是否加入背包
total_weight = 0
total_utility = 0
backpack_item = []
for i_item_sorted in range(item_num):
    if total_weight + item_weight[i_item_sorted] <= max_weight:
        total_weight += item_weight[i_item_sorted]
        total_utility += item_utility[i_item_sorted]
        backpack_item.append(item_index[i_item_sorted])

# 儲存演算法一結果
result = sorted(backpack_item)

# 演算法二：當下總效用最大法
# 根據效能大小，將各物品由效用大依序排列至效能小者
i_comparison = 0  # 初始值從效用列表的第一個值開始
while i_comparison < item_num - 1:  # 當比較到最後一個值時則停
    # 從自己的下一個開始比較到最後一個，在所有效用比自己還大的值中最靠近右邊的，與之交換位置(aka越小的要排到越右邊)
    index_to_exchange = i_comparison  # 交換對象的初始值為自己
    for i_following in range(i_comparison + 1, item_num):
        # 若比較對象的效用大於自己，應與之交換，將其索引值暫存
        if item_utility_algo2[i_comparison] < item_utility_algo2[i_following]:
            index_to_exchange = i_following
        # 若比較對象的效用等於自己的，則進一步判斷重量
        elif item_utility_algo2[i_comparison] == item_utility_algo2[i_following]:
            # 若比較對象的重量小於自己，應與之交換，將其索引值暫存
            if item_weight_algo2[i_comparison] > item_weight_algo2[i_following]:
                index_to_exchange = i_following
            # 若比較對象的重量等於自己的，則進一步判斷編號值
            elif item_weight_algo2[i_comparison] == item_weight_algo2[i_following]:
                # 若比較對象的編號比自己小，應與之交換，將其索引值暫存
                if item_index_algo2[i_comparison] > item_index_algo2[i_following]:
                    index_to_exchange = i_following
    # 當判斷出有交換時，進行位置交換
    if index_to_exchange != i_comparison:
        item_utility_algo2[i_comparison], item_utility_algo2[index_to_exchange] = item_utility_algo2[index_to_exchange], item_utility_algo2[i_comparison]
        item_weight_algo2[i_comparison], item_weight_algo2[index_to_exchange] = item_weight_algo2[index_to_exchange], item_weight_algo2[i_comparison]
        item_index_algo2[i_comparison], item_index_algo2[index_to_exchange] = item_index_algo2[index_to_exchange], item_index_algo2[i_comparison]
    # 當那個位置之後的效用值沒有再比自己更大時，表這個位置的乘客已經確定，下一個loop會開始確認下一個位置
    else:
        i_comparison += 1

# 根據剛剛排出的順序，依序判斷是否加入背包
total_weight_algo2 = 0
total_utility_algo2 = 0
backpack_item_algo2 = []
for i_item_sorted in range(item_num):
    if total_weight_algo2 + item_weight_algo2[i_item_sorted] <= max_weight:
        total_weight_algo2 += item_weight_algo2[i_item_sorted]
        total_utility_algo2 += item_utility_algo2[i_item_sorted]
        backpack_item_algo2.append(item_index_algo2[i_item_sorted])

# 儲存演算法二結果
result_algo2 = sorted(backpack_item_algo2)


# 選出總效用最大的那個結果並印出
if total_utility > total_utility_algo2:
    best_result = result
else:
    best_result = result_algo2

for i_item in range(len(best_result)):
    if (i_item != len(best_result) - 1):
        print(best_result[i_item], end=",")
    else:
        print(best_result[i_item])
```

```python
# 輸入一：投資標的個數、預算、風險趨避係數
targetnum_budget_coeff = input().split(',')
target_num = int(targetnum_budget_coeff[0])
budget = int(targetnum_budget_coeff[1])
coeff = int(targetnum_budget_coeff[2])

# 輸入二：資金需求
fund_needed = input().split(',')
for i_target in range(target_num):
    fund_needed[i_target] = int(fund_needed[i_target])

# 輸入三：預期報酬
expected_return = input().split(',')
for i_target in range(target_num):
    expected_return[i_target] = int(expected_return[i_target])

# 輸入四：風險共變數矩陣
risk_matrix = []
for i in range(target_num):
    # 輸入共變數列表
    individual_covariate = input().split(',')
    # 將值的型別轉換為integer
    for i_covariate in range(len(individual_covariate)):
        individual_covariate[i_covariate] = int(individual_covariate[i_covariate])
    # 將該列資料存入矩陣中
    risk_matrix.append(individual_covariate)

# 建立投資標的編號列表
index = []
for i_item in range(1, target_num + 1):
    index.append(i_item)

# 最終選擇的投資標的列表
portfolio = []

# 最終選擇的總資金需求與報酬
total_fund_needed = 0
total_expected_return = 0

# 演算法開始
last_turn_object_value = 0
continue_loop = True
while continue_loop:
    i_item = 0  # 演算法循環都是從第一個標的開始，故索引值設為0
    # 搜索還未被選入的投資標的
    tmp_max_object_index = -999
    tmp_max_object_fund_needed = 0
    tmp_max_object_value = 0
    for i_item in range(len(index)):
        # 若選擇該投資標的不會超過預算，則執行判斷
        if total_fund_needed + fund_needed[index[i_item] - 1] <= budget:
            # 暫時名單：把這個item加進去投資標的看看
            # 暫時名單內所儲存的值為初始輸入資料順序的索引值
            tmp_portfolio = []
            tmp_portfolio.append(index[i_item] - 1)  # 編號轉換成索引值，要減1
            for i_item_portfolio in portfolio:
                tmp_portfolio.append(i_item_portfolio - 1)  # 編號轉換成索引值，要減1

            # 計算risk
            risk = 0  # 初始化risk，避免累加到上一個loop的東西
            for i in tmp_portfolio:
                for j in tmp_portfolio:
                    risk += coeff*risk_matrix[i][j]

            # 計算return
            retrn = 0  # 初始化retrn，避免累加到上一個loop的東西
            for i in tmp_portfolio:
                retrn += expected_return[i]

            # 計算目標式值
            object_value = retrn - risk
            # 若目標式值相較上一輪的目標式值有增加，則進一步判斷
            if object_value > last_turn_object_value:
                if object_value > tmp_max_object_value:
                    tmp_max_object_index = i_item
                    tmp_max_object_fund_needed = fund_needed[index[i_item] - 1]
                    tmp_max_object_value = object_value
                elif object_value == tmp_max_object_value:
                    if fund_needed[index[i_item] - 1] < tmp_max_object_fund_needed:
                        tmp_max_object_index = i_item
                        tmp_max_object_fund_needed = fund_needed[index[i_item] - 1]
                        tmp_max_object_value = object_value
                    elif fund_needed[index[i_item] - 1] == tmp_max_object_fund_needed:
                        if index[i_item] < index[tmp_max_objective_index]:
                            tmp_max_object_index = i_item
                            tmp_max_object_fund_needed = fund_needed[index[i_item] - 1]
                            tmp_max_object_value = object_value

    # 若此次搜索有所斬獲，儲存資料
    if tmp_max_object_index != -999:
        total_fund_needed += tmp_max_object_fund_needed
        total_expected_return += expected_return[tmp_portfolio[0]]
        last_turn_object_value = tmp_max_object_value
        portfolio.append(index.pop(tmp_max_object_index))
    else:
        # 此次搜索已無新的進展
        continue_loop = False

portfolio.sort()
# 輸出結果
if len(portfolio) == 0:
    print(0)
else:
    for i_item in range(len(portfolio)):
        if i_item != len(portfolio) - 1:
            print(portfolio[i_item], end=",")
        else:
            print(portfolio[i_item])
```
